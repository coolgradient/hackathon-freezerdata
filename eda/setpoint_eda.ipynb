# Get the data

```python
import os
from pathlib import Path

project_dir = Path.cwd().parent
data_dir = project_dir / "data"
error_dir = data_dir / "Data Errors"
normal_dir = data_dir / "Data Normal"
setpoint_dir = data_dir / "Data Setpoints"
```

```python
import pandas as pd

df_setpoint = pd.read_csv(setpoint_dir / "22_05_2023-23_05_2023_merged.csv")
df_setpoint.sort_values(by="Time", inplace=True)
```

```python
for col in df_setpoint.columns:
    print(col)
```

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(
    rows=3, cols=1,
    shared_xaxes=True,
    vertical_spacing=0.05,
    subplot_titles=("Druk & Setpoints", "Fan Sturing","EnergyKWh")
)

# Row 1: druk & setpoints
fig.add_trace(
    go.Scatter(x=df_setpoint["Time"], y=df_setpoint["Condensatiedruk"], mode="lines", name="Condensatiedruk"),
    row=1, col=1
)
fig.add_trace(
    go.Scatter(x=df_setpoint["Time"], y=df_setpoint["Verdampingsdruk"], mode="lines", name="Verdampingsdruk"),
    row=1, col=1
)
fig.add_trace(
    go.Scatter(x=df_setpoint["Time"], y=df_setpoint["setpoint_min"], mode="lines", line=dict(dash="dash"), name="setpoint_min"),
    row=1, col=1
)
fig.add_trace(
    go.Scatter(x=df_setpoint["Time"], y=df_setpoint["setpoint_max"], mode="lines", line=dict(dash="dash"), name="setpoint_max"),
    row=1, col=1
)
fig.add_trace(
    go.Scatter(x=df_setpoint["Time"], y=df_setpoint["VerdamperFAN sturing"], mode="lines", name="VerdamperFAN sturing"),
    row=2, col=1
)
fig.add_trace(
    go.Scatter(x=df_setpoint["Time"], y=df_setpoint["CondensorFAN sturing"], mode="lines", name="CondensorFAN sturing"),
    row=2, col=1
)

# Row 2: EnergyKWh
fig.add_trace(
    go.Scatter(x=df_setpoint["Time"], y=df_setpoint["EnergyKWh"], mode="lines", name="EnergyKWh"),
    row=3, col=1
)

fig.update_layout(height=600, showlegend=True)
fig.show()
```
